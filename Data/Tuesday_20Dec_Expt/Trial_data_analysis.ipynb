```python
import sys, math
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter, medfilt 
import numpy as np
```

```python
def find_nearest(array,value):
    array = np.array(array)
    idx=(np.abs(array-value)).argmin()
    return idx
```

```python
def plot_data(datafile):
    #datafile = 'Test1/frames100Median5_Thresh50BlackPixels.txt'
    data = [[], [], [], [], [], [], []]
    ifile=open(datafile, 'rU')
    for line in ifile.readlines():
        #format is:
        '''
        Frame 0
        Black Pixels = 44003
        Hough Radius = 22
        Hough x = 
        Hough y = 
        Time :355521

        '''
        line=line.strip()
        if 'Frame' in line:
            line=line.split(' ')
            data[0].append(line[-1])
        elif 'Black' in line:
            line=line.split('=')
            data[2].append(line[-1])
        elif 'Time' in line:
            line=line.split(':')
            data[3].append(line[-1])
        elif 'Hough Radius' in line:
            line=line.split('=')
            data[4].append(line[-1])
        elif 'Hough x' in line:
            line=line.split('=')
            data[5].append(line[-1])
        elif 'Hough y' in line:
            line=line.split('=')
            data[6].append(line[-1])

    #m = max([float(i) for i in data[2]])
    m = 15000
    '''
    plt.figure(1, figsize=(12,6))
    plt.title(datafile)
    #plt.plot(data[0],data[1], 'r.-', label='No Median Blur')
    plt.plot(data[0],data[2], 'b.-', label='Median Blur: '+datafile.split('_')[3][-2:])
    #print(data)
    plt.ylim(0, m)
    plt.ylabel('Number of Black Pixels')
    plt.xlabel('Frame',fontsize=15)
    plt.legend(loc='best')
    plt.savefig(datafile+'_plot_frames.png')
    '''
    plt.figure(2, figsize=(12,6))
    plt.title(datafile)
    plt.plot([float(i)/1000000 for i in data[3]],data[2], 'k.-', label='Black Pixels')

    #we filter a filter using a Savitzky-Golay filter for non-periodic signals
    #using a polynomial regression over a small window of data [5 datapoint to order 1 polynomial]
    avg_blur = savgol_filter(data[2], 5 , 1)
    plt.plot([float(i)/1000000 for i in data[3]], avg_blur,'r--', label='No Blur filtered')

    var_blur = []

    #calculating variation measurement
    for i in range(len(data[2])):
        dp1 = (float(data[2][i]) - float(avg_blur[i]))**2
        var_blur.append(dp1)

    var_blur = math.sqrt(sum(var_blur)/(len(var_blur)-1))#calculate sqrt of sample variance (s^2)

    av_f_p_t = 'Average Frame Processing time: '+str(round(float(data[3][-1])/(len(data[3])*1000000),2))+' s/frame'
    av_fps = 'Average FPS: '+str(round(1000000*(float(data[0][-1])/float(data[3][-1])),2))
    ttl_t = 'Total time: '+str(round(float(data[3][-1])/1000000,2))+' s'
    b_variance = 'Variance: '+str(round(var_blur,2))

    print(av_f_p_t)
    print(av_fps)
    print(ttl_t)
    print(b_variance)
    
    #plt.text(0.45*float(data[3][-1])/1000000,0.1*m,av_f_p_t, fontsize=15)
    #plt.text(0.45*float(data[3][-1])/1000000,0.2*m,av_fps, fontsize=15)
    #plt.text(0.45*float(data[3][-1])/1000000,0.3*m,ttl_t, fontsize=15)
    #plt.text(0.45*float(data[3][-1])/1000000,0.4*m,b_variance, fontsize=15)

    
    
    #plt.ylim(0, m)
    plt.ylabel('Number of Black Pixels')
    plt.xlabel('Time / s',fontsize=15)
    #plt.legend(loc='best')
    #plt.savefig(datafile+'_plot_time.png')

    figure(3, figsize=(12,6))
    plt.plot([float(i)/1000000 for i in data[3]],data[4], 'b.-', label='Hough Radius')
    plt.ylabel('Hough Radius / px')
    plt.xlabel('Time / s',fontsize=15)
    #plt.legend(loc='best')
```

```python
def readin(filename):
    data = [[], [], [], [], [], [], []]
    ifile=open(filename, 'rU')
    for line in ifile.readlines():
        #format is:
        '''
        Frame 0
        Black Pixels = 44003
        Hough Radius = 22
        Hough x = 
        Hough y = 
        Time :355521

        '''
        line=line.strip()
        if 'Frame' in line:
            line=line.split(' ')
            data[0].append(int(line[-1]))
        elif 'Black' in line:
            line=line.split('=')
            data[2].append(int(line[-1]))
        elif 'Time' in line:
            line=line.split(':')
            data[3].append(float(line[-1]))
        elif 'Hough Radius' in line:
            line=line.split('=')
            data[4].append(float(line[-1]))
        elif 'Hough x' in line:
            line=line.split('=')
            data[5].append(int(line[-1]))
        elif 'Hough y' in line:
            line=line.split('=')
            data[6].append(int(line[-1]))
    dic = {
    'Frame':data[0],
    'Pixels':data[2],
    'Px_radius':[],
    'Time':data[3],
    'Hough Radius':data[4],
    #'Hough Centre':[(data[5][i],data[6][i]) for i in range(len(data[5]))]
    }
    
    for entry in dic['Pixels']:
        dic['Px_radius'].append(math.sqrt((entry/math.pi)))
    #print(dic)
    return dic
```

# Metrics to calculate are amplitude from baseline and time to half return to baseline and amplitude after 7 seconds

```python
def peak_metrics(example_dataset1, window_start, window_end):
    
    #inputs should be input file and times within which peaks sit
    
    
    #example_dataset1 = readin('5xKaps_Tests/Test8_frames400Median5_Thresh50BlackPixels.txt')

    figure(1, figsize=(20,10))
    s1 = find_nearest(example_dataset1['Time'], (window_start*1000000.0))
    e1 = find_nearest(example_dataset1['Time'], (window_end*1000000.0))
    
    plot([((float(i)/1000000)) for i in example_dataset1['Time']][s1:e1],example_dataset1['Px_radius'][s1:e1], linestyle='-', marker='.', color='r', label = 'Raw Data')

    y1 = medfilt(example_dataset1['Px_radius'][s1:e1],7) # median filter the signal

    plot([((float(i)/1000000)) for i in example_dataset1['Time']][s1:e1],y1, linestyle='-', color='b', label = 'Median Filtered')
    
    #plt.ylim(0, 15000)
    #plt.xlim(0, 15)
    ylabel('Effective Pupil Radius / px')
    xlabel('Time / s',fontsize=15)
    
    print('FPS')
    fps = float(len(example_dataset1['Time'][s1:e1]))/((max(example_dataset1['Time'][s1:e1]) - min(example_dataset1['Time'][s1:e1]))/1000000)
    print(fps)
    ##Baseline - first 4 seconds
    print('Baseline')
    base_window_start = window_start
    base_window_end = window_start+4.0
    baseline = sum(example_dataset1['Px_radius'][int(fps*base_window_start):int(fps*base_window_end)])/len(example_dataset1['Px_radius'][int(fps*base_window_start):int(fps*base_window_end)])
    print(baseline)
    
    axhline(y=baseline, linestyle='--', color='g')
    base_end_time = example_dataset1['Time'][int(fps*base_window_end)]*1/1000000.0

    axvline(x=base_end_time, linestyle='--', color='k')
    base_start_time = example_dataset1['Time'][int(fps*base_window_start)]*1/1000000.0

    axvline(x=base_start_time, linestyle='--', color='k')
    ##Amplitude
    print('Amplitude')
    p_start = int(fps*base_window_end)
    p_end = int(fps*window_end)
    minimum = min(example_dataset1['Px_radius'][p_start:p_end])
    amplitude = baseline - minimum 
    axhline(y=(baseline - amplitude), linestyle='--', color='k')
    print(amplitude)
    
    ##50% time
    print('50% height')
    half_max = amplitude/2.0
    axhline(y=(baseline - half_max), linestyle='--', color='r')
    print(half_max)
    
    #time between peak and 50%
    peak_time_index = example_dataset1['Px_radius'].index(minimum)
    peak_time = example_dataset1['Time'][peak_time_index]
    
    half_time = example_dataset1['Time'][find_nearest(example_dataset1['Px_radius'][peak_time_index:],(baseline-half_max))+peak_time_index]
    
    axvline(x=half_time/1000000.0, linestyle='--', color='r')
    print('50% time')
    print((half_time-peak_time)/1000000.0)
    
    ##Amplitude at t after peak
    print('Amplitude at [1,2,3,4,5,7] after peak')
    print()
    
    for i in [0,1,2,3,4,5,6,7]:
        tt = peak_time+i*1000000
        plot(tt/1000000, example_dataset1['Px_radius'][find_nearest(example_dataset1['Time'], tt)], 'ro')
    
    ##Amplitude at t after start
    peak_start = peak_time
    peak_time_index = example_dataset1['Time'].index(peak_time)
    for i in reversed(example_dataset1['Px_radius'][s1:peak_time_index]):
        #print(i)
        #check if within 10* of baseline
        peak_start = example_dataset1['Time'][example_dataset1['Px_radius'].index(i)]
        if (baseline - i) < 0.05*baseline:
            break
    
    for i in [0,1,2,3,4,5,6,7]:
        tt = peak_start+i*1000000
        plot(tt/1000000, example_dataset1['Px_radius'][find_nearest(example_dataset1['Time'], tt)], 'bo')
    
    print('Amplitude at [1,2,3,4,5,7] after start')
    print()
```
